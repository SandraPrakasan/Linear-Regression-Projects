```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
df=pd.read_csv("C:/Users/sandr/anaconda_projects/9e445923-85c6-4435-b2e3-d34a436af156/canada_per_capita_income.csv")
df
```

```python
year=df[["year"]]
pci=df[["per capita income (US$)"]]
plt.scatter(year,pci,color='red',marker='+')
```

```python
model=linear_model.LinearRegression()
model.fit(year,pci)
```

```python
test_data=pd.DataFrame({'year':[2017,2018,2019,2020]})
predicted_pci= model.predict(test_data)
```

```python
test_data['predicted per capita income'] = predicted_pci
test_data
```

```python
test_data.to_csv('per-capita-prediction-output.csv',mode='w')
```
