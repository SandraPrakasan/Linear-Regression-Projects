```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
table={'Area':[2600,3000,3200,3600,4000],'Price':[550000,565000,610000,680000,725000]}
df=pd.DataFrame(table)
df
```

```python
area=df[['Area']]
price=df[['Price']]
plt.plot(area,price,'r+')
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
model=linear_model.LinearRegression()
model.fit(area,price)
```

```python
new_data=pd.DataFrame({'Area':[3300,2800,3500,3786,2000,4400]})
predicted_price=model.predict(new_data)
for ip,op in zip(new_data['Area'],np.nditer(predicted_price)):
    print(f"{ip} : {op.round(2)}")
```

```python
model.coef_
```

```python
model.intercept_
```

```python
new_data['Predicted Price']=predicted_price
```

```python
new_data.to_csv('Predicted_Price.csv',mode='w')
```
